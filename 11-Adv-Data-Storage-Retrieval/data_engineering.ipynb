```python
# import dependencies
import numpy as np
import pandas as pd
import os
```

```python
# define paths of files
hawaii_measurements_csv = os.path.join('Resources', 'hawaii_measurements.csv')
hawaii_stations_csv = os.path.join('Resources', 'hawaii_stations.csv')
```

```python
# create dataframes with csv files
hi_measurements = pd.read_csv(hawaii_measurements_csv)
hi_stations = pd.read_csv(hawaii_stations_csv)
```

```python
# review measurements dataframe
print(f'there are {len(hi_measurements)} rows')
print(f'below are NaN counts by Series\n{hi_measurements.isnull().sum()}')

hi_measurements.head(10)
```

```python
# remove NaN rows from hi_measurements
hi_measurements_clean = hi_measurements.dropna()

# review new measurements dataframe
print(f'there are {len(hi_measurements_clean)} rows')
print(f'there are {hi_measurements_clean["prcp"].isnull().sum()} rows with in NaN values')

hi_measurements_clean.head(10)
```

```python
# save to csv
new_hawaii_measurements_csv = os.path.join('Resources', 'clean_hawaii_measurements.csv') 
hi_measurements_clean.to_csv(new_hawaii_measurements_csv)
```

```python
# review stations dataframe
print(f'there are {len(hi_stations)} rows')
print(f'below are NaN counts by Series\n{hi_stations.isnull().sum()}')

hi_stations.head(10)
```

```python
# save to csv
new_hawaii_stations_csv = os.path.join('Resources', 'clean_hawaii_stations.csv') 
hi_stations.to_csv(new_hawaii_stations_csv)
```
